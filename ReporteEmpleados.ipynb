```python
import pandas as pd

EmpleadosAttrition = pd.read_csv('/content/0.mikh5r1gs60.fukzkek4yb4empleadosRETO.csv')

if 'YearsAtCompany' not in EmpleadosAttrition.columns:
    EmpleadosAttrition['HiringDate'] = pd.to_datetime(EmpleadosAttrition['HiringDate'], errors='coerce')
    EmpleadosAttrition['HiringDate'] = EmpleadosAttrition['HiringDate'].dt.year
    EmpleadosAttrition['YearsAtCompany'] = 2018 - EmpleadosAttrition['HiringDate']

if 'DistanceFromHome_km' not in EmpleadosAttrition.columns and 'DistanceFromHome' in EmpleadosAttrition.columns:
    EmpleadosAttrition = EmpleadosAttrition.rename(columns={'DistanceFromHome': 'DistanceFromHome_km'})
if 'DistanceFromHome' not in EmpleadosAttrition.columns and 'DistanceFromHome_km' in EmpleadosAttrition.columns:
    EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome_km'].str.replace('km', '', regex=False)

columns_to_drop = ['HiringDate', 'Year', 'DistanceFromHome_km']
EmpleadosAttrition = EmpleadosAttrition.drop(columns=columns_to_drop, errors='ignore')

EmpleadosAttrition
```

```python
SueldoPromedioDepto = EmpleadosAttrition.groupby('Department')['MonthlyIncome'].mean().reset_index()
SueldoPromedioDepto = SueldoPromedioDepto.rename(columns={'MonthlyIncome': 'SueldoPromedio'})
SueldoPromedioDepto
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
columnas_categoricas = ['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Attrition']

for columna in columnas_categoricas:
    EmpleadosAttrition[columna] = le.fit_transform(EmpleadosAttrition[columna])

EmpleadosAttrition
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

EmpleadosAttrition_encoded = EmpleadosAttrition.copy()

le = LabelEncoder()

for column in EmpleadosAttrition_encoded.select_dtypes(include=['object']).columns:
    EmpleadosAttrition_encoded[column] = le.fit_transform(EmpleadosAttrition_encoded[column])
correlation_with_attrition = EmpleadosAttrition_encoded.corr()['Attrition'].sort_values(ascending=False)

correlation_with_attrition
```

```python
variables_to_keep = correlation_with_attrition[abs(correlation_with_attrition) >= 0.1].index

EmpleadosAttritionFinal = EmpleadosAttrition_encoded[variables_to_keep]
EmpleadoAttritionFinal
```
